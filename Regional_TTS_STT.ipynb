```python
import requests
from IPython.display import Audio
import base64
```

```python

from google.colab import userdata
SARVAM_API_KEY = userdata.get("sarvam_ai")
print(SARVAM_API_KEY)
```

 **TEXT-TO-SPEECH**

```python
url = "https://api.sarvam.ai/text-to-speech"

playload = {
    "inputs": ["बिलकुल! यहाँ एक साधारण हिंदी में लिखा हुआ संदेश है:नमस्ते! कैसे हैं आप? मुझे उम्मीद है कि आप अच्छे स्वास्थ्य में होंगे। अगर आपके पास कोई सवाल है या किसी चीज़ में मदद चाहिए, तो कृपया बेझिझक पूछें।धन्यवाद और शुभकामनाएँ!"],
    "target_language_code":"hi-IN",
    "speaker":"meera",
    "pitch":0,
    "pace":1.65,
    "loudness":2,
    "speech_sample_rate":8000,
    "enable_preprocessing":True,
    "model":"bulbul:v1"
}
headers = {"Content-Type": "application/json",
           "API-Subscription-Key":SARVAM_API_KEY}

response = requests.request("POST", url, json=playload, headers=headers)
```

```python
response.text
```

```python
audio_string = response.text[12:-3]
audio_data = base64.b64decode(audio_string)

with open("output.wav", "wb") as audio_file:
    audio_file.write(audio_data)
```

```python
audio_file = open("output.wav", "rb")
Audio(audio_file.read(), autoplay=True)
```

SPEECH-TO-TEXT

```python
url = "https://api.sarvam.ai/speech-to-text"

data ={
    "language_code":"hi-IN",
    "model":"saarika:v1",
}
files ={
    "file":("output.wav",open("output.wav","rb"),"audio/wav")
}
headers = {"API-Subscription-Key":SARVAM_API_KEY}
response = requests.post(url,data=data,files=files,headers=headers)
print(response.json())
```
